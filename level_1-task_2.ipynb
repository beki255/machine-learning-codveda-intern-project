### Import Libraries

```python
# Import required libraries for linear regression
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import warnings
warnings.filterwarnings('ignore')
print("Libraries imported successfully!")
```

### Load Dataset

```python
# Load house prediction dataset (space-separated values)
df = pd.read_csv('4) house Prediction Data Set.csv', sep='\s+')

print("Dataset shape:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

### Explore Dataset

```python
print("Dataset Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nBasic Statistics:")
print(df.describe())

print("\nColumn Names:")
print(df.columns.tolist())
```

### Identify Features and Target

```python
# Target is the last column (median house value)
target_column = df.columns[-1]
X = df.drop(target_column, axis=1)
y = df[target_column]

print(f"Target column: {target_column}")
print(f"Features shape: {X.shape}")
print(f"Target shape: {y.shape}")
print(f"\nTarget statistics:")
print(y.describe())
```

### Preprocess Data

```python
X_preprocessed = X.copy()
y_preprocessed = y.copy()

# Handle missing values for numerical columns
numerical_cols = X_preprocessed.select_dtypes(include=[np.number]).columns
for col in numerical_cols:
    if X_preprocessed[col].isnull().any():
        mean_val = X_preprocessed[col].mean()
        X_preprocessed[col].fillna(mean_val, inplace=True)
        print(f"Filled missing values in {col} with mean: {mean_val:.2f}")

# Handle missing values for categorical columns
categorical_cols = X_preprocessed.select_dtypes(include=['object']).columns
for col in categorical_cols:
    if X_preprocessed[col].isnull().any():
        mode_val = X_preprocessed[col].mode()[0]
        X_preprocessed[col].fillna(mode_val, inplace=True)
        print(f"Filled missing values in {col} with mode: {mode_val}")

print(f"\nMissing values after preprocessing: {X_preprocessed.isnull().sum().sum()}")
```

### Split Dataset into Training and Testing Sets

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_preprocessed, y_preprocessed,
    test_size=0.2,
    random_state=42
)

print("=== Dataset Split Results ===")
print(f"Total samples: {len(df)}")
print(f"Training set: {X_train.shape[0]} samples ({X_train.shape[0]/len(df)*100:.1f}%)")
print(f"Testing set: {X_test.shape[0]} samples ({X_test.shape[0]/len(df)*100:.1f}%)")
print(f"\nTraining features shape: {X_train.shape}")
print(f"Testing features shape: {X_test.shape}")
```

### Train Linear Regression Model

```python
model = LinearRegression()
model.fit(X_train, y_train)

print("=== Linear Regression Model Trained ===")
print(f"Intercept (β₀): {model.intercept_:.4f}")
print(f"Number of features: {len(model.coef_)}")
```

### Interpret Model Coefficients

```python
print("Model Coefficients (β₁, β₂, ...):")
for i, col in enumerate(X_train.columns):
    print(f"  {col}: {model.coef_[i]:.4f}")

print("\nInterpretation:")
print("- Positive coefficient: Increase in feature value increases predicted house price")
print("- Negative coefficient: Increase in feature value decreases predicted house price")
```

### Evaluate Model

```python
y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print("=== Model Evaluation ===")
print(f"R-squared (R²) on test set: {r2:.4f}")
print(f"Mean Squared Error (MSE) on test set: {mse:.4f}")
print("\nR² Interpretation: Proportion of variance in target explained by the model (closer to 1 is better)")
print("MSE Interpretation: Average squared difference between predicted and actual values (lower is better)")
```
